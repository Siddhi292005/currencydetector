```python
import os, shutil

eur_src = '/content/drive/MyDrive/train'
eur_dst = '/content/eur_sorted'

# EUR denomination mapping
eur_denoms = ['5', '10', '20', '50', '100', '200']

# Create destination folders
for d in eur_denoms:
    os.makedirs(os.path.join(eur_dst, f'EUR_{d}'), exist_ok=True)

# Sort images by denomination from filename
sorted_count = {}
skipped = []

for fname in os.listdir(eur_src):
    if not fname.lower().endswith(('.jpg', '.jpeg', '.png')):
        continue

    matched = False
    # Check filename prefix — order matters (check 100 before 10, 50 before 5)
    for denom in ['200', '100', '50', '20', '10', '5']:
        if fname.startswith(f'{denom}Euro') or fname.lower().startswith(f'{denom}euro'):
            dst_folder = os.path.join(eur_dst, f'EUR_{denom}')
            shutil.copy(os.path.join(eur_src, fname),
                        os.path.join(dst_folder, fname))
            sorted_count[f'EUR_{denom}'] = sorted_count.get(f'EUR_{denom}', 0) + 1
            matched = True
            break

    if not matched:
        skipped.append(fname)

print("=== EUR sorted counts ===")
for cls, count in sorted(sorted_count.items()):
    print(f"  {cls}: {count} images")

print(f"\nSkipped {len(skipped)} unmatched files")
if skipped:
    print("Sample skipped:", skipped[:5])
```

```python
import os, shutil, random

newdatapath = '/content/drive/MyDrive/Datasetsv2'
split_base  = '/content/drive/MyDrive/Datasetsv2_split'

# Clean and recreate
if os.path.exists(split_base):
    shutil.rmtree(split_base)
    print("Cleared old split folder")

random.seed(42)

inr_classes = ['INR_10', 'INR_20', 'INR_50', 'INR_100', 'INR_200', 'INR_500']

# Process one class at a time with confirmation
all_classes = sorted([
    c for c in os.listdir(newdatapath)
    if os.path.isdir(os.path.join(newdatapath, c)) and not c.startswith('.')
])

print(f"Found {len(all_classes)} classes: {all_classes}\n")

for cls in all_classes:
    cls_path = os.path.join(newdatapath, cls)

    all_images = [
        f for f in os.listdir(cls_path)
        if f.lower().endswith(('.jpg', '.jpeg', '.png'))
    ]

    train_dir = os.path.join(split_base, 'train', cls)
    val_dir   = os.path.join(split_base, 'val', cls)
    os.makedirs(train_dir, exist_ok=True)
    os.makedirs(val_dir, exist_ok=True)

    if cls in inr_classes:
        downloaded = [f for f in all_images if f.startswith(('inr_', 'kaggle_'))]
        original   = [f for f in all_images if not f.startswith(('inr_', 'kaggle_'))]
        random.shuffle(original)
        split      = int(len(original) * 0.8)
        orig_train = original[:split]
        orig_val   = original[split:]

        for f in downloaded + orig_train:
            shutil.copy(os.path.join(cls_path, f), os.path.join(train_dir, f))
        for f in orig_val:
            shutil.copy(os.path.join(cls_path, f), os.path.join(val_dir, f))

        print(f" {cls:<12} train={len(downloaded)+len(orig_train):<5} val={len(orig_val)}")

    else:
        random.shuffle(all_images)
        split      = int(len(all_images) * 0.8)
        train_imgs = all_images[:split]
        val_imgs   = all_images[split:]

        for f in train_imgs:
            shutil.copy(os.path.join(cls_path, f), os.path.join(train_dir, f))
        for f in val_imgs:
            shutil.copy(os.path.join(cls_path, f), os.path.join(val_dir, f))

        print(f"{cls:<12} train={len(train_imgs):<5} val={len(val_imgs)}")

print("\nAll classes split complete")
```

```python
import os
import numpy as np
import shutil
from sklearn.utils.class_weight import compute_class_weight
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout, BatchNormalization
from tensorflow.keras.applications import ResNet50V2
from tensorflow.keras.applications.resnet_v2 import preprocess_input
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.regularizers import l2

split_base = '/content/drive/MyDrive/Datasetsv2_split'
SAVE_DIR   = '/content/drive/MyDrive/currencydetector1'
```

```python
train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=15,
    zoom_range=0.2,
    width_shift_range=0.15,
    height_shift_range=0.15,
    shear_range=0.1,
    brightness_range=[0.8, 1.2],
    horizontal_flip=False,
    channel_shift_range=20.0,
    fill_mode='nearest'
)

val_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input
)

train_data = train_datagen.flow_from_directory(
    os.path.join(split_base, 'train'),
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    shuffle=True
)

val_data = val_datagen.flow_from_directory(
    os.path.join(split_base, 'val'),
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    shuffle=False
)

print("Classes:", train_data.class_indices)
print("Train samples:", train_data.samples)
print("Val samples:  ", val_data.samples)

NUM_CLASSES = len(train_data.class_indices)
assert NUM_CLASSES == 28, f"Expected 16, got {NUM_CLASSES}"
print(f"{NUM_CLASSES} classes confirmed")
```

```python
labels = train_data.classes
class_weights_array = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(labels),
    y=labels
)
class_weights = dict(enumerate(class_weights_array))
print("Class weights:")
for i, (cls, w) in enumerate(zip(train_data.class_indices.keys(), class_weights_array)):
    print(f"  {cls:<12} {w:.3f}")
```

```python
base = ResNet50V2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base.trainable = False

model = Sequential([
    base,
    GlobalAveragePooling2D(),
    BatchNormalization(),
    Dense(256, activation='relu', kernel_regularizer=l2(0.01)),
    Dropout(0.5),
    Dense(NUM_CLASSES, activation='softmax')
])

print("Output shape:", model.output_shape)
assert model.output_shape == (None, 28)
print("Model ready")
```

```python

callbacks = [
    EarlyStopping(
        monitor='val_accuracy', patience=5,
        restore_best_weights=True, verbose=1
    ),
    ReduceLROnPlateau(
        monitor='val_loss', factor=0.3,
        patience=3, min_lr=1e-8, verbose=1
    ),
    ModelCheckpoint(
        f'{SAVE_DIR}/best_multicurrency.keras',
        monitor='val_accuracy',
        save_best_only=True, verbose=1
    )
]
```

```python
print("\n=== Phase 1: Training head only ===")
model.compile(
    optimizer=Adam(learning_rate=1e-3),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history1 = model.fit(
    train_data,
    validation_data=val_data,
    epochs=10,
    class_weight=class_weights,
    callbacks=callbacks
)

model.save(f'{SAVE_DIR}/multicurrency_phase1.keras')
print(" Phase 1 saved")
```
