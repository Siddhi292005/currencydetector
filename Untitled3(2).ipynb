```python
import os
import numpy as np
import shutil
from sklearn.utils.class_weight import compute_class_weight
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout, BatchNormalization
from tensorflow.keras.applications import ResNet50V2
from tensorflow.keras.applications.resnet_v2 import preprocess_input
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

checkpoints_path = os.path.join(datapath, '.ipynb_checkpoints')
if os.path.exists(checkpoints_path):
    shutil.rmtree(checkpoints_path)
    print("Cleaned .ipynb_checkpoints")

classes = [f for f in os.listdir(datapath)
           if os.path.isdir(os.path.join(datapath, f))
           and not f.startswith('.')]
print(f"Classes found: {sorted(classes)}")
print(f"Total classes: {len(classes)}")
assert len(classes) == 7, f"Expected 7, got {len(classes)}"
print("Dataset clean")

#preprocessing and data augmentation
train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    validation_split=0.2,
    rotation_range=15,
    zoom_range=0.2,
    width_shift_range=0.15,
    height_shift_range=0.15,
    shear_range=0.1,
    brightness_range=[0.85, 1.15],
    horizontal_flip=False,
    channel_shift_range=20.0
)

val_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    validation_split=0.2
)
```

```python
 train_data = train_datagen.flow_from_directory(
    datapath,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    subset='training',
    shuffle=True
)
```

```python
val_data = val_datagen.flow_from_directory(
    datapath,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    subset='validation',
    shuffle=False
)


print("\nClass mapping:", train_data.class_indices)
print("Training samples:", train_data.samples)
print("Validation samples:", val_data.samples)
NUM_CLASSES = len(train_data.class_indices)
print("Number of classes:", NUM_CLASSES)
```

```python
labels = train_data.classes
class_weights_array = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(labels),
    y=labels
)
class_weights = dict(enumerate(class_weights_array))
print("\nClass weights:", class_weights)
```

```python

#base model
base = ResNet50V2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base.trainable = False

NUM_CLASSES = len(train_data.class_indices)
```

```python

#classification model

model = Sequential([
    base,
    GlobalAveragePooling2D(),
    BatchNormalization(),
    Dense(256, activation='relu'),
    Dropout(0.3),
    Dense(NUM_CLASSES, activation='softmax')
])
```

```python
callbacks = [
    EarlyStopping(
        monitor='val_accuracy', patience=5,
        restore_best_weights=True, verbose=1
    ),
    ReduceLROnPlateau(
        monitor='val_loss', factor=0.5,
        patience=3, min_lr=1e-7, verbose=1
    ),
    ModelCheckpoint(
        f'{SAVE_DIR}/best_checkpoint.keras',
        monitor='val_accuracy',
        save_best_only=True,
        verbose=1
    )
]
```

```python
print("\n=== Phase 1: Training head only (base frozen) ===")
model.compile(
    optimizer=Adam(learning_rate=1e-3),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history1 = model.fit(
    train_data,
    validation_data=val_data,
    epochs=10,
    class_weight=class_weights,
    callbacks=callbacks
)


model.save(f'{SAVE_DIR}/currex_phase1.keras')
print("Phase 1 saved to Drive")
```

```python
from tensorflow.keras.models import load_model
model1=load_model('/content/drive/MyDrive/currencydetector1/currex_phase1.keras')
print('phase1 loaded')
```

```python
print("\n=== Phase 2: Fine-tuning top 30 layers ===")
base.trainable = True
for layer in base.layers[:-30]:
    layer.trainable = False

model.compile(
    optimizer=Adam(learning_rate=1e-5),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history2 = model.fit(
    train_data,
    validation_data=val_data,
    epochs=25,
    class_weight=class_weights,
    callbacks=callbacks
)
#final accuracy evaluation
loss, acc = model.evaluate(val_data, verbose=0)
print(f"\nFinal val accuracy : {acc*100:.2f}%")
print(f"Final val loss     : {loss:.4f}")

model.save(f'{SAVE_DIR}/currex_phase2_val{acc*100:.0f}.keras')
print(f"Phase 2 saved as currex_phase2_val{acc*100:.0f}.keras")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report

val_data.reset()
y_pred_probs = model.predict(val_data, verbose=1)
y_pred = np.argmax(y_pred_probs, axis=1)
y_true = val_data.classes

class_labels = list(train_data.class_indices.keys())

print("\n=== Classification Report ===\n")
print(classification_report(y_true, y_pred, target_names=class_labels))

cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_labels, yticklabels=class_labels)
plt.title('Confusion Matrix — CURREX INR Model')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.tight_layout()
plt.savefig(f'{SAVE_DIR}/confusion_matrix.png', dpi=150)
plt.show()

print("\n=== Per-class accuracy ===")
for i, label in enumerate(class_labels):
    class_total = np.sum(y_true == i)
    class_correct = cm[i][i]
    print(f"  ₹{label:<6} {class_correct}/{class_total}  →  {class_correct/class_total*100:.1f}%")
```

```python
import shutil, os

base = '/content/kaggle_data/Indian currency dataset v1'

for split in ['training', 'validation']:
    split_path = os.path.join(base, split)
    for cls in ['10', '100', '20', '200', '2000', '50', '500']:
        src = os.path.join(split_path, cls)
        dst = os.path.join(datapath, cls)
        if not os.path.exists(src):
            print(f"Not found: {src}")
            continue
        copied = 0
        for fname in os.listdir(src):
            if fname.lower().endswith(('.jpg', '.jpeg', '.png')):
                shutil.copy(
                    os.path.join(src, fname),
                    os.path.join(dst, f"kaggle_{split}_{fname}")
                )
                copied += 1
        print(f"₹{cls} ({split}): copied {copied} images")
```
