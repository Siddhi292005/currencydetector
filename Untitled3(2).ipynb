```python


import os
import numpy as np
import shutil
import random
from sklearn.utils.class_weight import compute_class_weight
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout, BatchNormalization
from tensorflow.keras.applications import ResNet50V2
from tensorflow.keras.applications.resnet_v2 import preprocess_input
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.regularizers import l2

new_base = "/content/drive/MyDrive/Datasets_split"
```

```python
checkpoints_path = os.path.join(datapath, '.ipynb_checkpoints')
if os.path.exists(checkpoints_path):
    shutil.rmtree(checkpoints_path)
    print("Cleaned .ipynb_checkpoints")

classes = [f for f in os.listdir(datapath)
           if os.path.isdir(os.path.join(datapath, f))
           and not f.startswith('.')]
print(f"Classes found: {sorted(classes)}")
print(f"Total classes: {len(classes)}")
assert len(classes) == 7, f"Expected 7, got {len(classes)}"
print("Dataset clean")
```

```python
# Delete and recreate split folder cleanly
if os.path.exists(new_base):
    shutil.rmtree(new_base)
    print("Cleared old split folder")

random.seed(42)

for cls in sorted(os.listdir(datapath)):
    cls_path = os.path.join(datapath, cls)
    if not os.path.isdir(cls_path) or cls.startswith('.'):
        continue

    all_images = [
        f for f in os.listdir(cls_path)
        if f.lower().endswith(('.jpg', '.jpeg', '.png'))
    ]

    kaggle_imgs   = [f for f in all_images if f.startswith('kaggle_')]
    original_imgs = [f for f in all_images if not f.startswith('kaggle_')]

    random.shuffle(original_imgs)
    split      = int(len(original_imgs) * 0.8)
    orig_train = original_imgs[:split]
    orig_val   = original_imgs[split:]

    train_dir = os.path.join(new_base, 'train', cls)
    val_dir   = os.path.join(new_base, 'val', cls)
    os.makedirs(train_dir, exist_ok=True)
    os.makedirs(val_dir, exist_ok=True)

    for f in orig_train + kaggle_imgs:
        shutil.copy(os.path.join(cls_path, f),
                    os.path.join(train_dir, f))

    for f in orig_val:
        shutil.copy(os.path.join(cls_path, f),
                    os.path.join(val_dir, f))

    print(f"₹{cls}: train={len(orig_train)+len(kaggle_imgs)}, val={len(orig_val)}")

print("\n✅ Split complete")
```

```python
print("\n=== Final split counts ===")
for split in ['train', 'val']:
    print(f"\n{split}/")
    total = 0
    for cls in sorted(os.listdir(os.path.join(new_base, split))):
        cls_path = os.path.join(new_base, split, cls)
        if os.path.isdir(cls_path):
            count = len(os.listdir(cls_path))
            total += count
            print(f"  ₹{cls}: {count} images")
    print(f"  Total: {total}")
```

```python
train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=25,
    zoom_range=0.3,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    brightness_range=[0.6, 1.4],
    horizontal_flip=False,
    channel_shift_range=40.0,
    fill_mode='nearest'
)

train_data = train_datagen.flow_from_directory(
    os.path.join(new_base, 'train'),  # ← fixed path
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    shuffle=True
)

print("Train classes:", train_data.class_indices)
print("Train samples:", train_data.samples)
NUM_CLASSES = len(train_data.class_indices)
assert NUM_CLASSES == 7
print(" Train generator ready")
```

```python
val_path = os.path.join(new_base, 'val')  # ← defined here

class_names = sorted(os.listdir(val_path))
class_indices = {name: i for i, name in enumerate(class_names)}
print("Val class mapping:", class_indices)

# Must match train
assert class_indices == train_data.class_indices, " Class mismatch!"
print(" Classes match train")

# Collect image paths and labels
image_paths = []
labels = []

for cls in class_names:
    cls_path = os.path.join(val_path, cls)
    if not os.path.isdir(cls_path):
        continue
    for fname in os.listdir(cls_path):
        if fname.lower().endswith(('.jpg', '.jpeg', '.png')):
            image_paths.append(os.path.join(cls_path, fname))
            labels.append(class_indices[cls])

print(f"Total val images: {len(image_paths)}")

image_paths = np.array(image_paths)
labels = np.array(labels)

def load_and_preprocess(path, label):
    img = tf.io.read_file(path)
    img = tf.image.decode_jpeg(img, channels=3)
    img = tf.image.resize(img, [224, 224])
    img = tf.cast(img, tf.float32)
    img = preprocess_input(img)
    label = tf.one_hot(label, depth=7)
    return img, label

val_dataset = tf.data.Dataset.from_tensor_slices((image_paths, labels))
val_dataset = val_dataset.map(load_and_preprocess,
                               num_parallel_calls=tf.data.AUTOTUNE)
val_dataset = val_dataset.batch(32).prefetch(tf.data.AUTOTUNE)

# Verify one batch
for batch_x, batch_y in val_dataset.take(1):
    print(f"Batch shape: {batch_x.shape}")
    print(f"Labels shape: {batch_y.shape}")
    print("Val dataset working")
```

```python
labels = train_data.classes
class_weights_array = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(labels),
    y=labels
)
class_weights = dict(enumerate(class_weights_array))
print("\nClass weights:", class_weights)
```

```python

#base model
base = ResNet50V2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base.trainable = False

NUM_CLASSES = len(train_data.class_indices)
```

```python

#classification model
from tensorflow.keras.regularizers import l2
model = Sequential([
    base,
    GlobalAveragePooling2D(),
    BatchNormalization(),
    Dense(128, activation='relu', kernel_regularizer=l2(0.001)),
    Dropout(0.6),
    Dense(NUM_CLASSES, activation='softmax')
])
```

```python
callbacks = [
    EarlyStopping(
        monitor='val_accuracy', patience=4,
        restore_best_weights=True, verbose=1
    ),
    ReduceLROnPlateau(
        monitor='val_loss', factor=0.3,
        patience=3, min_lr=1e-8, verbose=1
    ),
    ModelCheckpoint(
        f'{SAVE_DIR}/best_checkpoint.keras',
        monitor='val_accuracy',
        save_best_only=True,
        verbose=1
    )
]
```

```python
print("\n=== Phase 1: Training head only (base frozen) ===")
model.compile(
    optimizer=Adam(learning_rate=1e-3),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history1 = model.fit(
    train_data,
    validation_data=val_dataset,
    epochs=10,
    class_weight=class_weights,
    callbacks=callbacks
)


model.save(f'{SAVE_DIR}/currex_phase1.keras')
print("Phase 1 saved to Drive")
```

```python
from tensorflow.keras.models import load_model
model1=load_model('/content/drive/MyDrive/currencydetector1/currex_phase1.keras')
print('phase1 loaded')
```

```python
print("\n=== Phase 2: Fine-tuning top 30 layers ===")
base.trainable = True
for layer in base.layers[:-10]:
    layer.trainable = False

model.compile(
    optimizer=Adam(learning_rate=1e-6),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history2 = model.fit(
    train_data,
    validation_data=val_dataset,
    epochs=15,
    class_weight=class_weights,
    callbacks=callbacks
)
#final accuracy evaluation
loss, acc = model.evaluate(val_dataset, verbose=0)
print(f"\nFinal val accuracy : {acc*100:.2f}%")
print(f"Final val loss     : {loss:.4f}")

model.save(f'{SAVE_DIR}/currex_phase2_val{acc*100:.0f}.keras')
print(f"Phase 2 saved as currex_phase2_val{acc*100:.0f}.keras")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report

# ── Rebuild true labels from val_path ─────────────────────────────
val_path = os.path.join(new_base, 'val')
class_names = sorted(os.listdir(val_path))
class_indices = {name: i for i, name in enumerate(class_names)}

val_image_paths = []
val_true_labels = []  # different variable name to avoid conflict

for cls in class_names:
    cls_path = os.path.join(val_path, cls)
    if not os.path.isdir(cls_path):
        continue
    for fname in sorted(os.listdir(cls_path)):
        if fname.lower().endswith(('.jpg', '.jpeg', '.png')):
            val_image_paths.append(os.path.join(cls_path, fname))
            val_true_labels.append(class_indices[cls])

val_true_labels = np.array(val_true_labels)
print(f"Val images: {len(val_true_labels)}")

# ── Get predictions ────────────────────────────────────────────────
y_pred_probs = model.predict(val_dataset, verbose=1)
y_pred = np.argmax(y_pred_probs, axis=1)

print(f"Predictions: {len(y_pred)}")
print(f"True labels: {len(val_true_labels)}")

# Trim to same length if needed
min_len = min(len(y_pred), len(val_true_labels))
y_pred = y_pred[:min_len]
y_true = val_true_labels[:min_len]

class_labels = list(train_data.class_indices.keys())

print("\n=== Classification Report ===\n")
print(classification_report(y_true, y_pred, target_names=class_labels))

cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_labels, yticklabels=class_labels)
plt.title('Confusion Matrix — CURREX INR Model')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.tight_layout()
plt.savefig(f'{SAVE_DIR}/confusion_matrix_v3.png', dpi=150)
plt.show()

print("\n=== Per-class accuracy ===")
for i, label in enumerate(class_labels):
    class_total = np.sum(y_true == i)
    class_correct = cm[i][i]
    print(f"  ₹{label:<6} {class_correct}/{class_total}  →  {class_correct/class_total*100:.1f}%")
```
